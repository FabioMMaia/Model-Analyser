```python
# Classification

from sklearn.datasets import make_classification

data =  make_classification(n_samples=2000, n_features= 30, n_informative=8, n_redundant=5,n_classes=4)

df = pd.DataFrame(data[0])
df['target'] = data[1]

m = df.shape[0]
n= df.shape[1]
```

```python
# Instiate object
FS = Feature_Selector(df, 'target')
```

```python
# Using logistic regression (default)
FS.classifier_univar_test(plot=True);
```
